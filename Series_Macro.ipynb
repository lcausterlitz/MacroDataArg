```python
import os
import sys

def install_if_needed(package):
    """Instala un paquete solo si no está instalado."""
    try:
        __import__(package)
    except ImportError:
        print(f"Instalando {package}...")
        !{sys.executable} -m pip install {package}

# Lista de librerías necesarias basadas en los imports proporcionados
required_libraries = [
    "selenium",
    "pyautogui",
    "xlwings",
    "pandas",
    "requests",
    "json",  # Nota: json es parte de la librería estándar de Python, no necesita instalación
    "datetime",  # Nota: datetime también es parte de la librería estándar, no necesita instalación
    "openpyxl",
    "xlsxwriter",
    "xlrd",
    "numpy",
    "scipy",
    "matplotlib",
    "warnings",  # Parte de la librería estándar, no necesita instalación
    "io",  # Parte de la librería estándar, no necesita instalación
    "yfinance"
]

# Instalar cada librería solo si no está instalada
for library in required_libraries:
    if library not in ["json", "datetime", "warnings", "io"]:  # Excluir las librerías estándar
        install_if_needed(library)

# Importar las librerías después de instalarlas (si fue necesario)

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options

import os
import time
import pyautogui
import xlwings as xw
import pandas as pd
import requests
import json
import datetime as dt
import openpyxl
import xlsxwriter
import xlrd
from pandas import DataFrame
import numpy as np
from numpy import matrix
from scipy import linalg
import matplotlib.pyplot as plt
import warnings
import io
import yfinance as yf
from datetime import date

# Marca de tiempo al inicio del programa
start_time = time.time()

```
